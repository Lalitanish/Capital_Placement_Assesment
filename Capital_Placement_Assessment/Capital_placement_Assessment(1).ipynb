Import Necessary Libraries:

The code starts by importing the required libraries: pdfplumber for extracting text from PDF files, os for file operations, and re for regular expressions.
Define Functions:

extract_text_from_pdf(pdf_path): This function takes a PDF file path, opens the PDF using pdfplumber, and extracts the text from each page. The extracted text is concatenated and returned.
extract_sections_from_text(pdf_text): This function takes extracted text from a PDF and uses regular expressions to extract three sections: experience, education, and skills.

The code defines the input folder path where the PDF files to be processed are located. This path needs to be adjusted to the correct folder path.

The code iterates through each file in the specified folder. If the file has a ".pdf" extension, it proceeds to process that PDF.
Extract Text and Sections:

For each PDF, it calls the extract_text_from_pdf function to extract the text from the PDF file.
It then calls the extract_sections_from_text function to extract the sections: experience, education, and skills, using regular expressions.
Print Extracted Sections:

The code prints the extracted sections for each PDF: experience, education, and skills.
The goal of this ** Pdf Extractor ** is to extract specific sections (experience, education, and skills) from PDF resumes and display them for further processing or analysis.
```python
#"C:\Users\DELL\Downloads\archive (7)\data\data\INFORMATION-TECHNOLOGY"
# This is PDF Extractor
!pip install pdfplumber
import pdfplumber
import os
import re

# Function to extract text from PDF file
def extract_text_from_pdf(pdf_path):
    pdf_text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            pdf_text += page.extract_text()

    return pdf_text

# Function to extract sections from PDF text
def extract_sections_from_text(pdf_text):
    # Regular expressions to match sections
    experience_pattern = r'Experience([\s\S]*?)Education'
    education_pattern = r'Education([\s\S]*?)Skills'
    skills_pattern = r'Skills([\s\S]*)'

    # Extract the sections using regex
    experience_match = re.search(experience_pattern, pdf_text)
    education_match = re.search(education_pattern, pdf_text)
    skills_match = re.search(skills_pattern, pdf_text)

    return (
        experience_match.group(1).strip() if experience_match else None,
        education_match.group(1).strip() if education_match else None,
        skills_match.group(1).strip() if skills_match else None
    )

# Input folder path
folder_path = "/content/drive/MyDrive/INFORMATION-TECHNOLOGY" # Replace with your folder path

# Iterate through PDF files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.pdf'):
        pdf_path = os.path.join(folder_path, filename)

        # Read the PDF and extract text
        pdf_text = extract_text_from_pdf(pdf_path)

        # Extract sections from PDF text
        experience_section, education_section, skills_section = extract_sections_from_text(pdf_text)

        # Print sections for each PDF
        print(f"File: {filename}")
        if experience_section:
            print("Experience:")
            print(experience_section)
        else:
            print("Experience section not found in the PDF.")

        if education_section:
            print("\nEducation:")
            print(education_section)
        else:
            print("Education section not found in the PDF.")

        if skills_section:
            print("\nSkills:")
            print(skills_section)
        else:
            print("Skills section not found in the PDF.")
        print("\n")
```

nlp = spacy.load("en_core_web_sm") loads the spaCy language model "en_core_web_sm," which is a pre-trained English language model for processing text.

tokenized_job_descriptions = [tokenize_job_description(desc) for desc in job_des_text] tokenizes each job description in the list job_des_text using the tokenize_job_description function.

```python
import pandas as pd
import spacy

# Load the spaCy language model
nlp = spacy.load("en_core_web_sm")

# Function to tokenize a job description using spaCy
def tokenize_job_description(job_desc):
    doc = nlp(job_desc)
    return [token.text for token in doc]

# Tokenize each job description
tokenized_job_descriptions = [tokenize_job_description(desc) for desc in job_des_text]

# Display the tokenized job descriptions
for i, tokens in enumerate(tokenized_job_descriptions):
    print(f'Tokenized job description {i+1}: {tokens}')

# In this code, we define a function tokenize_job_description to tokenize a single job description. We then iterate over the list of job descriptions, tokenizing each one using the tokenize_job_description function. Finally, we display the tokenized job descriptions.


```

Import Necessary Libraries:

The code starts by installing and importing the required libraries: scikit-learn for cosine similarity computation and CountVectorizer for text vectorization.
Define a Cosine Similarity Computation Function:

compute_cosine_similarity(tokens1, tokens2) is a function that computes cosine similarity between two sets of tokens. It first joins the tokens into text, then vectorizes and computes the cosine similarity.
Compute Cosine Similarity for Education and Skills:

The code calculates the cosine similarity between the tokenized education and each job description (tokenized_job_descriptions). It does the same for tokenized skills and each job description.
Combine Similarity Scores:

The similarity scores for education and skills are combined to obtain a single score for each job description. This combined score is the average of the education similarity score and the skills similarity score for each job description.
Create a DataFrame to Store the Results:

A DataFrame named result_df is created to store the job descriptions, education similarity scores, skills similarity scores, and combined similarity scores.
Rank the Job Descriptions Based on Combined Similarity:

The job descriptions are ranked based on the combined similarity score in descending order, and the top 5 are selected.
Print the Top 5 Matched Job Descriptions:

The code prints the top 5 job descriptions along with their education similarity, skills similarity, and combined similarity.
In summary, this code takes tokenized education and skills sections, computes cosine similarity scores between these sections and each job description, combines the scores, and ranks the job descriptions based on the combined similarity. Finally, it prints the top 5 matched job descriptions, providing insights into the relevance of job descriptions to the given education and skills sections.

```python
!pip install scikit-learn
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer


# Function to compute cosine similarity between two sets of tokens
def compute_cosine_similarity(tokens1, tokens2):
    text1 = ' '.join(tokens1)
    text2 = ' '.join(tokens2)
    vectorizer = CountVectorizer().fit([text1, text2])
    vectors = vectorizer.transform([text1, text2])
    similarity = cosine_similarity(vectors)[0, 1]
    return similarity

# Compute cosine similarity between tokenized_education and each job description
education_similarity_scores = [compute_cosine_similarity(tokenized_education, job_desc) for job_desc in tokenized_job_descriptions]

# Compute cosine similarity between tokenized_skills and each job description
skills_similarity_scores = [compute_cosine_similarity(tokenized_skills, job_desc) for job_desc in tokenized_job_descriptions]

# Combine similarity scores for education and skills
combined_similarity_scores = [(education_similarity_scores[i] + skills_similarity_scores[i]) / 2 for i in range(len(tokenized_job_descriptions))]

# Create a DataFrame to store the results
result_df = pd.DataFrame({
    'Job Description': [' '.join(tokens) for tokens in tokenized_job_descriptions],
    'Education Similarity': education_similarity_scores,
    'Skills Similarity': skills_similarity_scores,
    'Combined Similarity': combined_similarity_scores
})

# Rank the job descriptions based on combined similarity
ranked_df = result_df.sort_values(by='Combined Similarity', ascending=False).head(5)

# Print the top 5 matched job descriptions for education and skills
print("Top 5 Matched Job Descriptions:")
print(ranked_df[['Job Description', 'Education Similarity', 'Skills Similarity', 'Combined Similarity']])
```
